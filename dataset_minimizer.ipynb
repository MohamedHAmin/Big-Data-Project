```python
import json
```

```python
dataset_parent_dir = 'G:/Big Data/'
```

```python
def skip_white_space(file):
    while True:
        char = file.read(1)
        if char != '\n' and char != ' ' and char != '\t' and char != '\r':
            char_len = len(char.encode('utf-8'))
            file.seek(file.tell() - char_len)
            break
```

```python
def parse_one(file):
    json_str = ''
    json_str += file.read(1)
    count = 1
    while count != 0:
        char = file.read(1)
        json_str += char
        if char == '{':
            count += 1
        elif char == '}':
            count -= 1
    return json_str
```

```python
league_dataset_path = dataset_parent_dir + 'matches.json'
with open(league_dataset_path, 'r', encoding='utf-8') as file:
    skip_white_space(file)
    char = file.read(1)
    count = 0
    json_str = ''
    while count < 100:
        skip_white_space(file)
        json_str += parse_one(file)
        skip_white_space(file)
        char = file.read(1)
        json_str += char
        if char != ',':
            print("sad")
        count += 1
```

```python
json_str = '[' + json_str
json_str = json_str[:-1] + ']'
my_json = json.loads(json_str)
```
